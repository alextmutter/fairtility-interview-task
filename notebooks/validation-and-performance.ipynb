```python
# - Demonstrate accurcay of optimised vs. naive method
# - Number of windows processed over time
# - Computation time differences
```

```python
import pandas as pd
import csv
import os
import matplotlib.pyplot as plt
```

```python
CSV_DIR = "../code"
NAIVE_OUTPUTS = "naive_outputs.csv"
OPTIMISED_OUTPUTS = "optimised_outputs.csv"
N_REAL_VS_NEC = "naive_real_vs_necessary_computations.csv"
O_REAL_VS_NEC = "optimised_real_vs_necessary_computations.csv"
```

```python
### Combining total outputs

naive_df = pd.read_csv(os.path.join(CSV_DIR, NAIVE_OUTPUTS))
naive_df['process'] = ['naive']

optimised_df = pd.read_csv(os.path.join(CSV_DIR, OPTIMISED_OUTPUTS))
optimised_df['process'] = ['optimised']

df_outputs = pd.concat([naive_df, optimised_df], ignore_index=True)
df_outputs = df_outputs[['process', 'total_naive_inferences', 'total_processing_time', 'total_processes']]

df_outputs
```

```python
fig, ax = plt.subplots(figsize=(6, 2))  # Adjust size
ax.axis("tight")
ax.axis("off")
ax.table(cellText=df_outputs.values, colLabels=df_outputs.columns, cellLoc="center", loc="center")
plt.show()
```

```python
### Combining computations over time and display as graph

naive_df = pd.read_csv(os.path.join(CSV_DIR, N_REAL_VS_NEC))
naive_df.columns = ['naive_necessary_processes', 'naive_real_processes']
x_naive = naive_df['naive_necessary_processes']
y_naive = naive_df['naive_real_processes']

optimised_df = pd.read_csv(os.path.join(CSV_DIR, O_REAL_VS_NEC))
optimised_df.columns = ['optimised_necessary_processes', 'optimised_real_processes']
x_optimised = optimised_df['optimised_necessary_processes']
y_optimised = optimised_df['optimised_real_processes']

# Plot
plt.plot(x_naive, y_naive, marker="o", linestyle="-", label="Naive pipeline")
plt.plot(x_optimised, y_optimised, marker="s", linestyle="--", label="Optimised pipeline")

# Labels and legend
plt.xlabel("Necessary computations")
plt.ylabel("Real computations")
plt.title("Number of real compared to necessary computations for the naive vs. optimised pipeline")
plt.legend()
plt.grid(True)

# Show the graph
plt.show()
# plt.savefig('real_vs_necessary_computations.png', bbox_inches='tight')
```
